```python
import json
import requests
import pandas as pd
import numpy as np
import chardet
import re
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# 下载图片预测文件
url = "https://raw.githubusercontent.com/udacity/new-dand-advanced-china/master/"\
      "%E6%95%B0%E6%8D%AE%E6%B8%85%E6%B4%97/WeRateDogs%E9%A1%B9%E7%9B%AE/image-predictions.tsv"
response = requests.get(url)
filename = 'image_predictions.tsv'
with open(filename, 'wb') as file:
    file.write(response.content)
```

```python
# 确定文件编码
# 由输出结果{'encoding': 'ascii', 'confidence': 1.0, 'language': ''}
# 可以看出编码方式为 ascii

filename = 'tweet_json.txt'
with open(filename, 'rb') as f:
    rawdata = f.read()
    print(chardet.detect(rawdata))
```

```python
# 读取本地json数据(tweet_json.txt)
# UTF-8 包含了ascii，这里同样也可以选择UTF-8 来解码

filename = 'tweet_json.txt'
with open(filename,'r',encoding='UTF-8') as f:
    data = [json.loads(row) for row in f]
```

```python
tweet_rows = []
for entry in data:
    row = (entry['id'],entry['retweet_count'],entry['favorite_count'])
    tweet_rows.append(row)

tweet = pd.DataFrame(tweet_rows,columns=['tweet_id','retweet_count','favorite_count'])
```

```python
archive = pd.read_csv('twitter-archive-enhanced.csv')
predications = pd.read_csv('image_predictions.tsv',sep='\t')
```

#### 狗狗地位列数据单元格重新整理

##### define
这里再开启新的一列命名为stage，存储狗狗的stage

**第一次修正：程序修正**

***法1，较为保守的匹配方式***
> 原生匹配，比如doggo，只匹配doggo，多一个前缀或者后缀都不行

* 优点：十分精准的匹配
* 缺点：可能会跟丢其他stage，比如 doggos，其实也是doggo

> 具体方法
* stage单元格存储为一个字符串
    * 如果text中未找到stage，那么存储为缺失值（np.nan）
    * 如果text中只找到了一个stage，那么存储该 stage （默认为此tweet 中狗狗对应的stage）
    * 如果text中发现了多个stage，则stage是一个由狗狗地位通过逗号间隔的字符串
    
***法2，更为激进的匹配方式***
>派生匹配，比如floof，我们检索到的任何单词中，只要floof在其中，就认为匹配到了 floof

>>比如：floofer，floof，floofs，floofie 等都认为是floof

* 优点：可以匹配到派生词，增加stage匹配的数量

* 缺点：有可能引入其他非 stage 本意的单词匹配，比如 asdfloofxfaa （这是我随便写的一个）,会匹配到 floof

>具体方法
* 使用正则表达式来匹配
    * 对text列数据进行处理，全部转为小写
    * 使用findall，找出doggo|pupper|puppo|floof
    * 再将每个值通过逗号相连成带逗号间隔的字符串
    * 替换空值为np.nan


**第二次修正：手工修正**

* 之后对于出现的多个stage的列，下面可以看到有十多行，那么此时，直接手动查看这十多行对应的text，人为判断该行的stage到底为何stage，再修改这些stage，即可

```python
# 下面将会使用两种方式对text中的 stage 进行一个抽取
# archive_clean_2 将会提供给第二种方法匹配用
archive_clean_2 = archive_clean.copy(deep=True)
```

##### code

###### 第一次修正，法1（原生匹配，较为保守的方式）

```python
# 保守方式会跟丢 doggo 这个stage
archive.text[83]
```

```python
stage_list = ['doggo', 'floofer', 'pupper', 'puppo']
stage = []

for row in archive_clean['text']:
    temp = []
    row = row.lower()
    
    # 将所有非字母字符全部替换为空格，然后再根据空格分割字符串，再来检验
    words = re.sub(r'[^a-z]',' ',row).split() 
    for word in stage_list:
        if word in words:
            temp.append(word)
    temp = ','.join(set(temp))
    temp = temp if temp else np.nan
    stage.append(temp)
archive_clean['stage'] = stage

archive_clean['stage'].value_counts()
```

##### 第一次修正，法2（派生匹配，更为激进的方式）

```python
archive_clean_2['stage']=archive_clean_2.text.str.lower().str.findall('(doggo|pupper|puppo|floof)')
archive_clean_2['stage'] = archive_clean_2['stage'].apply(lambda x: ','.join(set(x)))
archive_clean_2['stage'].replace('', np.nan, inplace = True)

archive_clean_2['stage'].value_counts()
```

##### 分割线

以上方法在一开始并未考虑到数据的质量问题，这里改用另一种方式重新提取分数，在上面，已经知道了评分的形式如下

* 单个评分
    * 带有小数点的评分
    * 多个狗狗只有一个评分：99/90
* 多个评分
    * text 中有多个评分
    
怎么提取？

* 编写正则表达式，来匹配单个评分类型的评分

* 使用findall 找出多个评分的行，查看具体text，人工审核

* 发现可以将带有多个评分的行的分数取平均值，合并至单个评分

* 查找以上未匹配到的行进行修正

* 将结果覆盖至archive_clean 的评分分子分母中

##### code3

```python
# 可以匹配到带小数点的分子
# 分母，可以匹配到10的倍数
# 注意，对于多组符合规律的评分，将只匹配第一组
rating = archive_clean.text.str.extract('((?:\d+\.)?\d+)\/([1-9]+[0]+)', expand = True)
rating.columns = ['rating_numerator', 'rating_denominator']
rating['rating_numerator'] = rating['rating_numerator'].astype(float)
rating['rating_denominator'] = rating['rating_denominator'].astype(float)
rating.info()
```

```python
# 提取每行所有 分子/分母 这种格式的数据至 rating_multi
rating_multi = archive_clean.text.str.findall('((?:\d+\.)?\d+\/[1-9]+[0]+)')
```

```python
# 审核多行评分对应的index
check_score_index = rating_multi[rating_multi.str.len()>1].index
print(len(check_score_index))
```

可以看出，共有24行是存在多个评分的，这里需要进行人工核查

```python
# 人工审核
for e in check_score_index:
    print(e)
    print(archive_clean.loc[e,'text'])
    print('*'*20)
```

```python
# 这里除了1165行有问题，其他均可以
rating_multi[check_score_index]
```

```python
# 上述24行数据中，我发现除了 index=1165 这行数据，其他行的评分都可以平均值来替代，一定程度上可以代表该tweet评分
# 多组评分，求取一个平均值来代替评分，再将评分进行分数化即可还原成 分子/分母 形式
rating_multi_avg = rating_multi[check_score_index].apply(lambda x:sum(eval(e) for e in x)/len(x))
rating_multi_avg
```

```python
# 分子圆整
rating_multi_numerator = rating_multi_avg.apply(lambda x:round(x*10))

# 修正index=1165 的分子
rating_multi_numerator[1165] = 13

rating_multi_numerator
```

```python
# 分母全部取10
rating_multi_denominator = rating_multi_avg.apply(lambda x:10)
rating_multi_denominator
```

```python
rating_multi_index = list(rating_multi_denominator.index)
rating_multi_index
```

```python
# 多组评分打包成元组
zip_multi_rating = zip(rating_multi_index, rating_multi_numerator, rating_multi_denominator)
for x,y,z in zip_multi_rating:
    print(x,y,z)    
```

```python
# 修正多组评分行数据
# 这里是直接覆盖到 rating 对应的index中即可
zip_multi_rating = zip(rating_multi_index, rating_multi_numerator, rating_multi_denominator)
for x,y,z in zip_multi_rating:
    rating.loc[x,'rating_numerator'] = y
    rating.loc[x,'rating_denominator'] = z
```

#### 狗狗的名字需要修正

##### define

**step1:提取name**

***正则表达式***
* 狗狗只有名，没有姓
    * 也就是说，只需要匹配一个名即可
    * 大写开头
    * 前面有特定语句搭配，引出狗狗name
        * ***这里我只考虑以下几种匹配情况***
            * 只有一个名字
                * This is dog_name
                * H/his name is dog_name
                * H/her name is dog_name
                * Meet dog_name
                * Say hello to dog_name
                * named dog_name
                * ...

            * 不止一个名字，这里我默认为两个名字
                * This is dog_name1 and dog_name2
                * Meet dog_name1 and dog_name2
                * Say hello to dog_name1 and dog_name2
                * Here we have dog_name1 and dog_name2
                * ...

***怎么提取狗狗名字***
* 按照上述规律编写好正则表达式

    * 我将以上三类合并为 空格 + is + 空格 + dog_name(大写开头)
        * This is dog_name
        * H/his name is dog_name
        * H/her name is dog_name
    * 另外，我发现狗狗的名字中有时会出现带有变音符号的字母
        * 比如index = 1559 对应的狗狗名字 Frönq
        * 这里需要使用 \w 来匹配余下部分，同时flags=re.U

* 对 archive_clean.text 使用 extract，提取狗狗名字


**step2:合并name**
* 对于不止一个狗狗名字的行来说，会找到2个狗狗名字，将这两个狗狗名字合并至一个单元格内，使用逗号间隔符分开
    * （这里我按照以上规律所写的正则表达式最多只能匹配到2个狗狗名字）
    * ***怎么合并？***
        * 类似 df.name1.map(str) + ',' + df.name2.map(str)
        * 如果name2 为np.nan,那么合并后，使用 replace('nan',''),再使用strip(',')即可去掉多余逗号
        * 再填充为空的单元格，使用 np.where(name,name,np.nan)

预测表格（predications），其中包含了预测结果的前三名，推特 ID，图像 url 以及最可信的预测结果对应的图像编号

**如何筛选？**
* 先看排名第一的预测结果，如果其中的判断是否是狗为真（True），那么我会将此记下，如果第一预测结果不为狗狗，那么再看第二预测结果，以此类推

* 如果三个预测对应的狗狗预测全为False，那么，将此行数据对应记为，未识别出狗狗



注意：这里需要创建的列（之后将此创建的数据合并至 archive 中）
* 预测的狗狗种类，使用***type***表示
* 预测的可信概率，使用***conf***表示

**我想要合并的方式**

* (p1_dog,p2_dog,p3_dog) 合并为 ***isdog***
    * 若为 True ，使用 1 指代
    * 若为 False，使用 0 指代
        * 例：结构格式为 101 
        * 表示 p1_dog=True, p2_dog=False, p3_dog=True
    
* (p1_conf,p2_conf,p3_conf) 合并为***conf***
* (p1,p2,p3) 合并为***type***


**如何将预测结果添加至 archive_clean中？**
* 使用pd.merge(archive_clean,pre_df,on='tweet_id',how='left')

##### code

```python
pre_backup = predications_clean.copy(deep=True)

# 置换True为1，False为0
pre_backup.replace({True:1,False:0},inplace=True)
```

```python
pre_backup['isdog'] = pre_backup.p1_dog.astype(str) + pre_backup.p2_dog.astype(str) + pre_backup.p3_dog.astype(str)
# pre_backup['isdog'] = pre_backup.p1_dog.map(str) + pre_backup.p2_dog.map(str) + pre_backup.p3_dog.map(str)
pre_backup['conf'] = pre_backup.p1_conf.astype(str) + ','+ pre_backup.p2_conf.astype(str) + ','+ pre_backup.p3_conf.astype(str)
pre_backup['type'] = pre_backup.p1.astype(str) + ','+ pre_backup.p2.astype(str)+ ','+ pre_backup.p3.astype(str)
pre_backup.head()
```

```python
# 构造confidence,dog_type列数据
confidence = []
dog_type =[]
for i in range(pre_backup.shape[0]):
    x = pre_backup.iloc[i].isdog.find('1')
    if x!= -1:
        confidence.append(pre_backup.iloc[i].conf.split(',')[x])
        dog_type.append(pre_backup.iloc[i].type.split(',')[x])
    else:
        confidence.append(0)
        dog_type.append('NotFound')
```

```python
id_type_conf = pd.DataFrame({'tweet_id':pre_backup.tweet_id,
                             'jpg_url':pre_backup.jpg_url,
                             'img_num':pre_backup.img_num,
                             'dog_type':dog_type,
                             'confidence':confidence},
                             columns = ['tweet_id','jpg_url','img_num','dog_type','confidence',])
id_type_conf.confidence = id_type_conf.confidence.astype(float)

# 合并
archive_clean = pd.merge(archive_clean, id_type_conf,on='tweet_id',how='left')

# 丢弃pre_backup中的p1,p1_conf,p1_dog,p2,p2_conf,p2_dog,p3,p3_conf,p3_dog 列
# 用另一种方式选取列也可达到丢弃列效果
pre_backup = pre_backup[['tweet_id','jpg_url','img_num','isdog','conf','type']]
```

* 评分（这里指分子/分母的比值）与 喜欢和转发数之间关系相关性虽然没有没有转发数和喜爱数之间的那么强烈，但是也是较高的，分别为 0.411237 和 0.308543

**评分与转发数的相关性较低原因，我的理解**
* 人与人之间的差异性和相似性造就了这个结果
    * 差异性，对狗狗评分的不同，这里其实可以通过查看一些tweet的评论发现，多数人的评分都各不一样。
        * 转发tweet的人跟推主的评分可能不一直，比如，同一张，图片，推主评分 12/10，那么有的人可能较为喜欢的14/10，低一些的10/10，都有可能
    * 相似性，对狗狗某些特征的有共同的喜爱，推主的评分代表了一部分人的评分，这里我称之为相似。
           
    * 具体点说：人们转发是出于自己的喜爱，而喜爱，是因人而异的
    
        * 这里做个思想实验，假设人们的喜爱同推主的一致，那么，根据上面喜爱和转发数的关系，相关系数高达0.91，再假设评分反应了推主的喜爱程度，那么可以推断出，评分越高，喜欢的程度就越高，转发的概率也就更高，也就表明，评分和转发相关性很强。现在只改变一个条件，假设人们的喜爱不大一致，那么，推主的评分高的话，只能代表推主的评分，不能代表全体人们内心的评分，反应在相关性上，比之前弱了些，也就是相关系数要更小些。
